```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from scipy.ndimage import rotate
from scipy.ndimage import gaussian_filter

# Load the image
Anime = 'AnimeRifai.jpg'
img = mpimg.imread(Anime)

# Figure for displaying original and annotated images
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
plt.imshow(img)
plt.axis('off')
plt.title('Original Image')

# Annotated Image
plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
plt.imshow(img)
plt.axis('off')
plt.annotate('Rifai', xy=(200, 200), xytext=(200, 200), 
             arrowprops=dict(facecolor='red', shrink=0.5), fontsize=14, color='red')
plt.title('Annotated Image')
plt.tight_layout()
plt.show()

# Figure for displaying rotated and filtered images
plt.figure(figsize=(12, 6))

# Rotated Image
rotated_img = rotate(img, 45, reshape=True)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
plt.imshow(rotated_img)
plt.axis('off')
plt.title('Rotated Image')

# Filtered Image
filtered_img = gaussian_filter(img, sigma=5)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
plt.imshow(filtered_img)
plt.axis('off')
plt.title('Filtered Image')

plt.tight_layout()
plt.show()

# Save the original image with a title as "Modul6"
plt.figure(figsize=(6, 6))
plt.imshow(img)
plt.axis('off')
plt.title('Saved Original Image')
plt.savefig('Modul6.png')
plt.show()
```

Pilihlah 2 gambar dari komputer anda untuk membuat sebuah overlay image.
Buka gambar pertama sebagai latar belakang dan gambar kedua yang akan disisipkan (overlay)
Untuk gambar kedua kalian bisa melakukan resize() terlebih dahulu (optional) agar gambar 2 yang menjadi overlay tidak lebih besar dari background image, kalian bisa membuat skala (100,100) pixel.
Sisipkan gambar overlay ke dalam gambar background menggunakan fungsi paste()
Simpan gambar hasil edit
Tampilkan hasil edit kepada asisten
 * hint : gunakanlah fungsi overlay/merging images setelah itu gunakan fungsi resize dan simpan gambar

```python
from PIL import Image

# TODO 1: Buka gambar
background_image = Image.open("image2.jpg")  # Ganti dengan path gambar latar belakang Anda
overlay_image = Image.open("image1.jpg")  # Ganti dengan path gambar overlay Anda

# TODO 2: Konversi overlay ke mode RGB (menghilangkan alpha channel jika ada)
overlay_image = overlay_image.convert("RGB")

# TODO 3: (Optional) Perkecil ukuran gambar overlay menggunakan resize()
overlay_image = overlay_image.resize((200, 200))  

# Hitung posisi pusat agar overlay ditampilkan di tengah
bg_width, bg_height = background_image.size
ov_width, ov_height = overlay_image.size
x_center = (bg_width - ov_width) // 2
y_center = (bg_height - ov_height) // 2

# TODO 4: Sisipkan gambar overlay ke dalam gambar background
background_image.paste(overlay_image, (x_center, y_center))

# TODO 5: Simpan gambar hasil edit
background_image.save("image3.jpg")

# TODO 6: Tampilkan gambar hasil edit
background_image.show()
```
