# data sistem informasi mata kuliah dalam bentuk tuple, dimana setiap tuple berisi informasi ‘mata_kuliah’, ‘average_nilai’, ‘jumlah_mahasiswa’

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data_matkul = [
    ("Fungsional", 87.5, 48),
    ("Jarkom", 91, 52),
    ("Mobile", 79.7, 40),
    ("Pirdas", 92.5, 49),
    ("Web", 89, 55)
]

# Step 1: Ekstrak data menggunakan list comprehension
mata_kuliah = [item[0] for item in data_matkul]
rata_rata = [item[1] for item in data_matkul]
jumlah_mahasiswa = [item[2] for item in data_matkul]

# Step 2: Hitung total nilai menggunakan map dan lambda
total_nilai = list(map(lambda x: x[1] * x[2], data_matkul))

# Menampilkan nama mata kuliah, rata-rata nilai, dan jumlah mahasiswa
print("Nama Mata Kuliah | Rata-rata Nilai | Jumlah Mahasiswa")
for i in range(len(mata_kuliah)):
    print(f"{mata_kuliah[i]:<15} | {rata_rata[i]:<15} | {jumlah_mahasiswa[i]:<15}")

# Step 3: Scatter plot untuk rata-rata nilai vs jumlah mahasiswa
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.scatter(rata_rata, jumlah_mahasiswa, color='blue', alpha=0.7)
plt.title("Scatter Plot: Rata-rata Nilai dan Jumlah Mahasiswa")
plt.xlabel("Rata-rata Nilai")
plt.ylabel("Jumlah Mahasiswa")
plt.grid(True)


# Step 4: Diagram batang untuk total nilai setiap mata kuliah
colors = ['blue', 'orange', 'green', 'red', 'purple'] 

plt.subplot(2, 2, 2)
plt.bar(mata_kuliah, total_nilai, color=colors, alpha=0.7)
plt.title("Diagram Batang: Total Nilai Setiap Mata Kuliah")
plt.xlabel("Mata Kuliah")
plt.ylabel("Total Nilai")


# Step 5: Diagram garis untuk rata-rata nilai tiap mata kuliah
plt.subplot(2, 2, 3)
plt.plot(mata_kuliah, rata_rata, marker='o', color='purple', linestyle='-', linewidth=2)
plt.title("Diagram Garis: Rata-rata Nilai Tiap Mata Kuliah")
plt.xlabel("Mata Kuliah")
plt.ylabel("Rata-rata Nilai")

# Step 6: Pie chart untuk jumlah mahasiswa
plt.subplot(2, 2, 4)
explode = [0.1 if jumlah == min(jumlah_mahasiswa) else 0 for jumlah in jumlah_mahasiswa]  # Explode data terkecil
plt.pie(jumlah_mahasiswa, labels=mata_kuliah, autopct='%1.1f%%', explode=explode, startangle=140, shadow= True)
plt.title("Pie Chart: Jumlah Mahasiswa Tiap Mata Kuliah")

# Step 7: Menampilkan semua plot dengan subplot
plt.tight_layout()
plt.show()
```

```python
#7

import pandas as pd

# Mendefinisikan list DESTINATIONS
DESTINATIONS = [
    ("Jakarta", "Surabaya", 120),
    ("Jakarta", "Bali", 150),
    ("Surabaya", "Medan", 200),
    ("Medan", "Jakarta", 180),
    ("Bali", "Surabaya", 170)
]

# Membuat dictionary dari DESTINATIONS
distance_dict = {
    f"{src} to {dest}": distance for src, dest, distance in DESTINATIONS
}

# Membuat Series
distance_series = pd.Series(distance_dict)
print(distance_series)
```

```python
#8

import pandas as pd

# Data DESTINATIONS
DESTINATIONS = [
    ("Jakarta", "Surabaya", 120),
    ("Jakarta", "Bali", 150),
    ("Surabaya", "Medan", 200),
    ("Medan", "Jakarta", 180),
    ("Bali", "Surabaya", 170)
]

# Membuat dictionary dari DESTINATIONS
destinations_dict = {
    "Asal": [src for src, _, _ in DESTINATIONS],
    "Destination": [dest for _, dest, _ in DESTINATIONS],
    "($)": [distance for _, _, distance in DESTINATIONS]
}

# Membuat DataFrame
destinations_df = pd.DataFrame(destinations_dict)
print(destinations_df)
```
